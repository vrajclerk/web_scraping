```python
import wikipediaapi
import requests
import csv

def get_wikipedia_page(title):
    user_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    wiki_wiki = wikipediaapi.Wikipedia('en', headers={'User-Agent': user_agent})
    page = wiki_wiki.page(title)
    
    if page.exists():
        # description = page.summary[:500]  # Limiting the summary to 500 characters
        photo_url = get_wikipedia_image(title)
        
        return {
            'name': page.title,
            'photo': photo_url
        }
    else:
        return None

def get_wikipedia_image(title):
    url = 'https://en.wikipedia.org/w/api.php'
    params = {
        'action': 'query',
        'titles': title,
        'prop': 'pageimages',
        'format': 'json',
        'pithumbsize': 500
    }
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
    response = requests.get(url, params=params, headers=headers)
    data = response.json()
    
    pages = data['query']['pages']
    for page_id, page_data in pages.items():
        if 'thumbnail' in page_data:
            return page_data['thumbnail']['source']
    return None

# List of Indian historical personalities
historical_personalities = [
    'Abdul Kalam',
    'Abdul Ghaffar Khan',
    'Amartya Sen',
    'Aryabhata',
    'Bhagat Singh',
    'Bhimrao Ambedkar',
    'Chandra Shekhar Azad',
    'Chhatrapati Shivaji Maharaj',
    'Dhirubhai Ambani',
    'Dnyaneshwar',
    'Eknath',
    'Fazlur Rahman',
    'Indira Gandhi',
    'Mahatma Gandhi',
    'Gauri Ma',
    'Gopal Krishna Gokhale',
    'Hemu',
    'Iqbal',
    'Jawaharlal Nehru',
    'Jitendra Nath Das',
    'Jodha Bai',
    'Kailash Satyarthi',
    'Kalpana Chawla',
    'Karna',
    'Kasturba Gandhi',
    'Lal Bahadur Shastri',
    'Lalita Pawar',
    'Madhav Rao I',
    'Madhav Rao II',
    'Mahesh Yogi',
    'Mangal Pandey',
    'Mira Bai',
    'Kublai Khan',
    'Morarji Desai',
    'Nandini Satpathy',
    'Motilal Nehru',
    'Netaji Subhas Chandra Bose',
    'Nita Ambani',
    'P. V. Narasimha Rao',
    'Pankaj Advani',
    'Rani Lakshmibai',
    'Rajiv Gandhi',
    'Ram Manohar Lohia',
    'Srinivasa Ramanujan',
    'Ratan Tata',
    'Sardar Vallabhbhai Patel',
    'Satyajit Ray',
    'Vinayak Damodar Savarkar',
    'Shivaji Maharaj',
    'Sitaram Yechury',
    'Swami Dayananda Saraswati',
    'Swami Vivekananda',
    'Rabindranath Tagore',
    'J. R. D. Tata',
    'Tukaram',
    'Ustad Bismillah Khan',
    'Vikram Sarabhai',
    'Vishwanath Pratap Singh',
    'Wamanrao D. S. Muktibodh',
    'Zakir Hussain',
    'Azim Premji',
    'Baba Amte',
    'Baba Ramdev',
    'Mahavir',
    'Chandra Sekhar',
    'Chinmayananda',
    'E. Sreedharan',
    'Firoz Shah Tughlaq',
    'Hema Malini',
    'Jayaprakash Narayan',
    'Kailash Chandra',
    'Mayawati',
    'Lala Lajpat Rai',
    'Patanjali',
    'Mokshagundam Visvesvaraya',
    'Ravi Shankar',
    'Rudyard Kipling',
    'Shiv Nadar',
    'Subramanian Swamy',
    'Veer Savarkar',
    'Zafar Ali Khan',
    'Zubaida Tariq'
]


# Fetch data for each personality
data = []
for personality in historical_personalities:
    try:
        page_data = get_wikipedia_page(personality)
        if page_data:
            data.append(page_data)
        else:
            print(f"Page for {personality} does not exist.")
    except Exception as e:
        print(f"Error fetching data for {personality}: {e}")

# Write data to a CSV file
with open('indian_historical_personalities1.csv', 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['name', 'photo']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    
    writer.writeheader()
    for entry in data:
        writer.writerow(entry)

print("Data has been written to indian_historical_personalities1.csv")
```
