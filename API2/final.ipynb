```python
import requests
from bs4 import BeautifulSoup

def get_image_urls(url):
    try:
        # Send a GET request to the URL
        response = requests.get(url)
        response.raise_for_status()  # Raise an exception for HTTP errors

        # Parse the HTML content
        soup = BeautifulSoup(response.content, 'html.parser')

        # Find all image elements
        img_tags = soup.find_all('img')

        # Extract the 'src' attribute of the first five images
        img_urls = [img['src'] for img in img_tags[:9] if 'src' in img.attrs]

        # Handle relative URLs
        img_urls = [requests.compat.urljoin(url, img_url) for img_url in img_urls]

        return img_urls
    except requests.exceptions.RequestException as e:
        print(f"Error fetching the URL: {e}")
        return []


name = input("Enter the Name: ").split(" ")
url = f"https://www.google.com/search?q={"+".join(name)}&sca_esv=045202e8b9d90fc1&tbm=isch&sxsrf=ADLYWIL331qKbpAaq_G_y4d0ALlfi7Yi8Q%3A1722071303407&source=hp&biw=1536&bih=695&ei=B7mkZs-hFtjt1e8PlKa_mAc&iflsig=AL9hbdgAAAAAZqTHF08-huRWLXOg5J8q4LtAs08GSp0u&oq=&gs_lp=EgNpbWciACoCCAAyBxAjGCcY6gIyBxAjGCcY6gIyBxAjGCcY6gIyBxAjGCcY6gIyBxAjGCcY6gIyBxAjGCcY6gIyBxAjGCcY6gIyBxAjGCcY6gIyBxAjGCcY6gIyBxAjGCcY6gJIvwhQAFgAcAF4AJABAJgBAKABAKoBALgBA8gBAIoCC2d3cy13aXotaW1nmAIBoAIGqAIKmAMGkgcBMaAHAA&sclient=img"
image_urls = get_image_urls(url)
print(image_urls)
```

```python
import os
import requests
import cv2

def download_image(folder_name, img_name, img_url, rootfolder='HDimages'):
    if not os.path.exists(rootfolder):
        os.makedirs(rootfolder)
    if not os.path.exists(os.path.join(rootfolder, folder_name)):
        os.makedirs(os.path.join(rootfolder, folder_name))
        
    img_response = requests.get(img_url)
    folder = os.path.join(rootfolder, folder_name)
    img_path = os.path.join(folder, img_name)
    
    with open(img_path, 'wb') as file:
        file.write(img_response.content)
    
    return img_path

def upscale_image(img_path, scale=2):
    # Load the image
    img = cv2.imread(img_path)
    
    # Initialize the model
    sr = cv2.dnn_superres.DnnSuperResImpl_create()
    
    # Read the model
    path = "EDSR_x2.pb"  # You need to download the model
    sr.readModel(path)
    
    # Set the model and scale
    sr.setModel("edsr", scale)
    
    # Upscale the image
    result = sr.upsample(img)
    
    # Save the upscaled image
    upscaled_img_path = img_path.replace(".jpg", f"_upscaled_x{scale}.jpg")
    cv2.imwrite(upscaled_img_path, result)
    
    return upscaled_img_path

# Example usage
# folder_name = "example_folder"
# img_name = "example_image.jpg"
# img_url = "https://example.com/high_resolution_image.jpg"

img_path = download_image(folder_name, img_name, img_url)
upscaled_img_path = upscale_image(img_path, scale=2)

print(f"Original image saved at: {img_path}")
print(f"Upscaled image saved at: {upscaled_img_path}")
```
