```python
import wikipediaapi
import requests
import csv

def get_wikipedia_page(title):
    user_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    wiki_wiki = wikipediaapi.Wikipedia('en', headers={'User-Agent': user_agent})
    page = wiki_wiki.page(title)
    
    if page.exists():
        # description = page.summary[:500]  # Limiting the summary to 500 characters
        photo_url = get_wikipedia_image(title)
        
        return {
            'name': page.title,
            'photo': photo_url
        }
    else:
        return None

def get_wikipedia_image(title):
    url = 'https://en.wikipedia.org/w/api.php'
    params = {
        'action': 'query',
        'titles': title,
        'prop': 'pageimages',
        'format': 'json',
        'pithumbsize': 500
    }
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
    response = requests.get(url, params=params, headers=headers)
    data = response.json()
    
    pages = data['query']['pages']
    for page_id, page_data in pages.items():
        if 'thumbnail' in page_data:
            return page_data['thumbnail']['source']
    return None

# List of Indian historical personalities
historical_personalities = [
    "A. P. J. Abdul Kalam",
    "Akhbar",
    "Arjuna",
    "Ashoka",
    "Bhagat Singh",
    "Bhimrao Ambedkar",
    "Chandragupta Maurya",
    "Chandra Sekhara Azad",
    "Chandragupta II",
    "Chaucer",
    "Devendra Nath Tagore",
    "Durga Das",
    "Gandhi",
    "Guru Nanak",
    "Hemu",
    "Jawaharlal Nehru",
    "Jodha Bai",
    "Julius Caesar",
    "Kailash Satyarthi",
    "Kalidasa",
    "Karna",
    "Keshav Chandra Sen",
    "Lal Bahadur Shastri",
    "Laxmi Bai",
    "Mahatma Gandhi",
    "Manu",
    "Mangal Pandey",
    "Moti Lal Nehru",
    "Nehru",
    "Netaji Subhas Chandra Bose",
    "P. V. Narasimha Rao",
    "Patel",
    "Rabindranath Tagore",
    "Rani Durgavati",
    "Rani Laxmibai",
    "Ratan Tata",
    "Ravi Shankar",
    "Rohit Vemula",
    "Sardar Patel",
    "Satyajit Ray",
    "Shivaji Maharaj",
    "Subhas Chandra Bose",
    "Swami Dayananda Saraswati",
    "Swami Vivekananda",
    "Tansen",
    "Vikramaditya",
    "Vikram Sarabhai",
    "Vishnu Sharma",
    "Vivekananda",
    "Zakir Husain",
    "Satyendranath Tagore",
    "Surya Sen",
    "Sundar Pichai",
    "P. Chidambaram",
    "Rajendra Prasad",
    "Rajiv Gandhi",
    "Rajkumar",
    "Ravi Shankar Prasad",
    "V. S. Naipaul",
    "Kiran Bedi",
    "Udit Narayan",
    "K. R. Narayanan",
    "M. S. Swaminathan",
    "M. F. Husain",
    "Srinivasa Ramanujan",
    "Sukumar Ray",
    "Indira Gandhi",
    "Raj Kumar",
    "K. Kamaraj",
    "M. K. Gandhi",
    "Ram Manohar Lohia",
    "C. N. Annadurai",
    "M. S. Golwalkar",
    "Swami Sivananda",
    "S. Radhakrishnan",
    "Shankaracharya",
    "Suresh Raina",
    "Satguru",
    "Dhirubhai Ambani",
    "K. R. Narayanan",
    "Vikram Sarabhai",
    "Gopal Krishna Gokhale",
    "Amitabh Bachchan"
]



# Fetch data for each personality
data = []
for personality in historical_personalities:
    try:
        page_data = get_wikipedia_page(personality)
        if page_data:
            data.append(page_data)
        else:
            print(f"Page for {personality} does not exist.")
    except Exception as e:
        print(f"Error fetching data for {personality}: {e}")

# Write data to a CSV file
with open('indian_historical_personalities1.csv', 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['name', 'photo']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    
    writer.writeheader()
    for entry in data:
        writer.writerow(entry)

print("Data has been written to indian_historical_personalities1.csv")
```
